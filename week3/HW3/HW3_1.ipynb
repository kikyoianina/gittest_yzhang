```python
from __future__ import print_function
import pylab as pl
import numpy as np
%pylab inline
```

```python
## Generate different sample sizes
size=2000/(np.array(range(2,102)))
print (size, size.shape)
```

```python
md = {'normal':{}, 'poisson':{}, 'binomial':{}, 'chisq':{}, 'geometric':{}, 'logistic':{}}
mymean = 100 ## Population mean
for n in size:
    md['normal'][n] = np.random.normal(loc=mymean, size=n)
    md['poisson'][n] = np.random.poisson(lam=mymean, size=n)
    md['binomial'][n] = np.random.binomial(mymean/0.5, 0.5, size=n)
    md['chisq'][n] = np.random.chisquare(mymean, size=n)
    md['geometric'][n] = np.random.geometric(1.0/mymean, size=n)
    md['logistic'][n] = np.random.logistic(loc=mymean, size=n)
means = {'normal':{}, 'poisson':{}, 'binomial':{}, 'chisq':{}, 'geometric':{}, 'logistic':{}}
```

```python
fig_mu_N = pl.figure(figsize=(30,20))

normal = fig_mu_N.add_subplot(231)
for i in md['normal'].iterkeys():
    means['normal'][i] = md['normal'][i].mean()
    normal.plot(i, means['normal'][i], 'o')
    normal.set_xlabel('sample size')
    normal.set_ylabel('sample mean')
    normal.plot([min(size), max(size)],[mymean,mymean], 'k')
    plt.title('Normal')
    
poisson = fig_mu_N.add_subplot(232)
for i in md['poisson'].iterkeys():
    means['poisson'][i] = md['poisson'][i].mean()
    poisson.plot(i, means['poisson'][i], 'o')
    poisson.set_xlabel('sample size')
    poisson.set_ylabel('sample mean')
    poisson.plot([min(size), max(size)],[mymean,mymean], 'k')
    plt.title('Poisson')
    
binomial = fig_mu_N.add_subplot(233)
for i in md['binomial'].iterkeys():
    means['binomial'][i] = md['binomial'][i].mean()
    binomial.plot(i, means['binomial'][i], 'o')
    binomial.set_xlabel('sample size')
    binomial.set_ylabel('sample mean')
    binomial.plot([min(size), max(size)],[mymean,mymean], 'k')
    plt.title('Binomial')

chisq = fig_mu_N.add_subplot(234)
for i in md['chisq'].iterkeys():
    means['chisq'][i] = md['chisq'][i].mean()
    chisq.plot(i, means['chisq'][i], 'o')
    chisq.set_xlabel('sample size')
    chisq.set_ylabel('sample mean')
    chisq.plot([min(size), max(size)],[mymean,mymean], 'k')
    plt.title('Chisquare')

geometric = fig_mu_N.add_subplot(235)
for i in md['geometric'].iterkeys():
    means['geometric'][i] = md['geometric'][i].mean()
    geometric.plot(i, means['geometric'][i], 'o')
    geometric.set_xlabel('sample size')
    geometric.set_ylabel('sample mean')
    geometric.plot([min(size), max(size)],[mymean,mymean], 'k')
    plt.title('Geometric')

logistic = fig_mu_N.add_subplot(236)
for i in md['logistic'].iterkeys():
    means['logistic'][i] = md['logistic'][i].mean()
    logistic.plot(i, means['logistic'][i], 'o')
    logistic.set_xlabel('sample size')
    logistic.set_ylabel('sample mean')
    logistic.plot([min(size), max(size)],[mymean,mymean], 'k')
    plt.title('Logistic')
```

## Comment
The 6 plots above look as we expected that most sample means are nearby 100, which is the population mean that we set. Also, we could observe that as the sample size increases, the sample mean becomes more and more close to the polulation mean, due to the reason that more outliers might be generated in small-size sample compared to large-size sample.


```python
allmeans=[]
for n1 in means:
    for n2 in means[n1]:
        allmeans.append(means[n1][n2])
pl.figure(figsize=(10,10))
pl.hist(allmeans,bins=30)
pl.xlabel('sample mean')
pl.ylabel('N')
```
